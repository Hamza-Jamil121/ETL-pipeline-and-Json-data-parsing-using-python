# Here i write sql query using sqldf package and perfrom data joining and data transformation task as per requirement

```python
#USE LIMIT KEYWORD TO READ ONLY 1000 ROWS, DUE TO THE LAPTOP PERFORMANCE ISSUES
from pandasql import sqldf
output = sqldf('''
with debit_ct as (select user_id,created_at,sum(amount) as amount_of_debit,
                  count(*) as no_of_debit_trxn   from data1 where transaction_type='debit' group by user_id,created_at),
                         credit_ct as (select  user_id,created_at,sum(amount) as amount_of_credit,
                  count(*) as no_of_credit_trxn   from data1 where transaction_type='credit' group by user_id,created_at),
                  
                  table2 as (select * from data2_v),
                  table3 as (select * from data3_v),
                  table1 as (select user_id,created_month,count(*) as no_of_transactions,sum(amount) as amount_of_total_transactions from data1
                  group by created_month,user_id),
                  
                   montly as (select created_month,sum(amount) as calculated_fields_months_transacting,
                   sum(customer_net_balance_after_transaction) as calculated_fields_median_gmv_per_month,
                   max(amount) as calculated_fields_median_trans_per_month


                   from data1
                  group by created_month)
                         
                         
    select  distinct
    db.user_id,
    c.amount_of_credit,
    db.amount_of_debit,
    tb1.amount_of_total_transactions
    ,c.no_of_credit_trxn
    ,db.no_of_debit_trxn,
    tb1.no_of_transactions,
    tb2.rating,
    tb2.user_last_activity,
    JULIANDAY(DATE()) - JULIANDAY(user_last_activity) as days_since_last_activity,
    db.created_at,
    JULIANDAY(DATE()) - JULIANDAY(tb2.user_signup_seconds_to_date) as days_since_signup,
    tb3.device_language as language,
    tb3.city,
    tb3.device_model as phone_model,
    tb3.app_version,
    mn.calculated_fields_median_gmv_per_month,
    mn.calculated_fields_median_trans_per_month,
    mn.calculated_fields_months_transacting
     
    from table1 tb1
    inner join  debit_ct db on tb1.user_id = db.user_id 
    inner join  credit_ct c on tb1.user_id = c.user_id 
    inner join table2 tb2 on tb1.user_id =tb2.id
    inner join table3 tb3 on tb1.user_id = tb3.user_id 
    inner join montly mn on   mn.created_month =tb1.created_month limit 1000

   
    
 
   
    


''')
    
   
  
```

# Create Sql connection using python to perfrom ETL Task and here we move data from final output table to Sql table

```python
import pyodbc
import json
server = "DESKTOP-G87QUGF"
db = 'test'
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server + ';DATABASE=' +
                      db + ';Trusted_Connection=yes')
cursor = conn.cursor()
cursor
```

# code for create table on sql database 

```python
# once table is created on sql just comment the code


cursor.execute('''
  CREATE TABLE etl_tablee (
   
[user_id]  nvarchar(1000),
amount_of_credit decimal(18,4),
amount_of_debit decimal(18,4),
amount_of_total_transactions decimal(18,4),
no_of_credit_trxn int,
no_of_debit_trxn int,
no_of_transactions int,
rating  nvarchar(100),
user_last_activity  nvarchar(100),
days_since_last_activity int, 
created_at  nvarchar(100),
days_since_signup  int,
[language]  nvarchar(100),
city  nvarchar(100),
phone_model  nvarchar(100), 
app_version  nvarchar(100),
calculated_fields_median_gmv_per_month decimal(18,4),
calculated_fields_median_trans_per_month decimal(18,4),
calculated_fields_months_transacting decimal(18,4)
      )
               ''')
```

```python
truncate_query = """ Truncate table  [etl_tablee] """
cursor.execute(truncate_query)
```

# data insertion from final output to SQL TABLE into etl_tablee

```python
for row in output.itertuples(truncate_query):
    cursor.execute('''
                INSERT INTO etl_tablee (
user_id,
amount_of_credit,
amount_of_debit,
amount_of_total_transactions,
no_of_credit_trxn,
no_of_debit_trxn,
no_of_transactions,
rating,
user_last_activity,
days_since_last_activity,
created_at,
days_since_signup,
language,
city,
phone_model,
app_version,
calculated_fields_median_gmv_per_month,
calculated_fields_median_trans_per_month,
calculated_fields_months_transacting)
                VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
                ''',
row.user_id,
row.amount_of_credit,
row.amount_of_debit,
row.amount_of_total_transactions,
row.no_of_credit_trxn,
row.no_of_debit_trxn,
row.no_of_transactions,
row.rating,
row.user_last_activity,
row.days_since_last_activity,
row.created_at,
row.days_since_signup,
row.language,
row.city,
row.phone_model,
row.app_version,
row.calculated_fields_median_gmv_per_month,
row.calculated_fields_median_trans_per_month,
row.calculated_fields_months_transacting
                )
conn.commit()
```
